```python
import pandas as pd
```

```python
# Load your weekly data
weekly_data = pd.read_csv('../data/featured-data.csv')
weekly_data['timestamp'] = pd.to_datetime(weekly_data['timestamp'])

print("üìä LSTM Feasibility Check\n" + "="*50)

# Check per-athlete data availability
athlete_weeks = weekly_data.groupby('athlete').size()

print(f"\n‚úÖ Total athletes: {len(athlete_weeks)}")
print(f"‚úÖ Total weeks: {len(weekly_data)}")
print(f"\nüìà Weeks per athlete:")
print(athlete_weeks.describe())

# Check how many athletes have enough data for LSTM (need at least 7 weeks)
sufficient_data = athlete_weeks[athlete_weeks >= 7]
print(f"\n‚úÖ Athletes with ‚â•7 weeks (can use for LSTM): {len(sufficient_data)}")
print(f"   That's {len(sufficient_data)/len(athlete_weeks)*100:.1f}% of athletes")

# Check for gaps in data
print(f"\nüìÖ Date range: {weekly_data['timestamp'].min()} to {weekly_data['timestamp'].max()}")

# Show sample athlete timeline
sample_athlete = weekly_data['athlete'].iloc[0]
sample_data = weekly_data[weekly_data['athlete'] == sample_athlete].sort_values('timestamp')
print(f"\nüìã Sample athlete ({sample_athlete}) timeline:")
print(sample_data[['timestamp', 'weekly_mileage', 'fatigue_index', 'avg_weekly_pace_km']].head(10))

print("\n" + "="*50)
if len(sufficient_data) > 50:
    print("‚úÖ YOU HAVE ENOUGH DATA FOR LSTM!")
    print(f"   Can create ~{len(sufficient_data) * (athlete_weeks.mean() - 6):.0f} training sequences")
else:
    print("‚ö†Ô∏è Limited data for LSTM (need more athletes with longer histories)")
```
